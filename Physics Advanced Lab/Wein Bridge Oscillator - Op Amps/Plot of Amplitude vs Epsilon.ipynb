```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

#libraries for reading csv
import csv
import pandas as pd
```

```python
#reading in data from /data .csv
data = pd.read_csv("data/opampData.csv")

df = pd.DataFrame(data)

epsilon = list(df["(Resistance - Onset) / Onset (no units)"])
epsilon_error = []

amplitude = list(df["Pk-Pk Amp / 2 (V)"])
amplitude_error = list(df["Error (V).1"])

#print(df)
#print(epsilon)
#print(amplitude)
#print(amplitude_error)
```

```python
#plotting the log/log plot using matplotlib
x = np.array(epsilon)
y = np.array(amplitude)

#fitting the plot
m, b = np.polyfit(np.log(x, where=x>0), np.log(y), 1)
fit = (m*x)+b
print(m, b, fit)

#creating figure and axes
fig, ax1 = plt.subplots(1, 1, figsize=[10, 8])

#plotting the graph with Log ticks at x and y axis using loglog
ax1.loglog(x, y, 'b', linewidth=2, label='Function')
ax1.loglog(x, fit, 'g--', linewidth=2, label='Fit')
ax1.set_title('Amplitude vs Epsilon for Wein Bridge Op-Amp Circuit', fontsize=20)
ax1.set_xlabel('Epsilon=(Gain-Onset)/Onset', fontsize=15)
ax1.set_ylabel('Peak Amplitude (V)', fontsize=15)
ax1.yaxis.grid(True, which='minor', linestyle='--')
#ax1.legend(loc=2, prop={'size':18})

plt.tight_layout()
plt.show()
```
