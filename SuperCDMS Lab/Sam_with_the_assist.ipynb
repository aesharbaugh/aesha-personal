````python
def ERFC(E, E_n, A3):
    """
    Generate the parameterized error function given input energies from
    response func -- refer to docstring for that function.
    
    NOTE :> Sometimes you'll want to split these one-liners out as an easy optimization.
            Depending on which Python (e.g. Cython/Jython) these can be implicitly 
            converted to lambda functions which will save on function stack usage, 
            especially when you are using multiple chained expressions in numpy. 
    
    You can explicitly make this a lambda function using the variable lambda 
    syntax in Python >=3.7.
    """
    # do explit casting here for np.float64, which should be valid for inputs and 
    # will be accepted by special.erfc
    return (0.5 * (special.erfc( 1 - ((E_n + 764 - E) / A3) )))
    
````

````python
def response_func(E, E_n, A):
    """
    Generate a response given input energies and statistical parameters
    for the stat. mech. system; for gaseous He-3. 
        
    Assumes use of numpy arrays with representation of underlaying datatype
    for numerical stability and array casting (i.e., call this function with
    an array of values input for @e and @e_n, using dtype=np.float64). 
    
    Params:
    @E    - the input energy for a given thingy (don't know what thingy)
    @E_n  - the input energy for a given neutron bound to the system? Related to e.
    @A    - a list of 18+1 entries (0th index null) governing the statistical distribution
    
    Returns:
    @response - the statistical response based on a stat. dist. of the input energies
                and background system information (params list A).
    """
    # Perform explicit type conversions, just to be absolutely sure about numerical stability
    # and underflow concerns. 
    E = np.float128(E)
    E_n = np.float128(E)
    A = np.array(A, dtype=np.float128)
    
    delta_flag = (E <= (E_n + 764.0))
    
    res = np.dtype(np.float128) # Change to np.float128 for precision. 
                                # Used float32 for numerical stability testing
                                # Find machine-dependent precision with ```np.finfo(np.float128).eps```
                                # np.finfo(np.float128).minexp \approx 10^-16382
      
    # begin adding term by term (each line); the gaussian terms
    res = (A[17] * (E - A[18]) + A[16]) / (1.0 + np.exp( (E - A[18])/A[15]) ) +\
          (A[11] * (E - A[13]) + A[12]) / (1.0 + np.exp( (E - A[13])/A[14]) )
    
    # add in parameterized error functions using coerced datatypes back to float64
    ## NOTE :> Not sure from image algo. desc. whether or not it is `np.exp(...) * ERFC` or if
    ##         ERFC should be included in the exponent (e.g. `np.exp(... * ERFC)`)
    res = res +\
          (A[6] * np.exp( -1.0 * ((E_n + 764.0 - E) / A[7])) * ERFC(np.float64(E), np.float64(E_n), np.float64(A[3]))) +\
          ((A[1] * A[4]) * np.exp( -1.0 * ((E_n + 764.0 - E) / A[5])) * ERFC(np.float64(E), np.float64(E_n), np.float64(A[3])))
    
    # Use delta flag switch to avoid unnecessary multiplication and error in mantissa. In older papers, they would
    # include flag variables like this as a mathematical formalism, but it's important to translate irl when implementing.
    # Unfortunately, this disables broadcasting and vectorization of the computation :/
    # NOTE :> Make sure to clarify what's supposed to be squared! (e.g. is the expr supposed to be 
    #         `np.exp(...)^2` or is it supposed to be `np.exp( (...)^2 )`?
    if (delta_flag):
        res = res + (A[1] * np.exp(-0.5 * np.power(((E_n + 764.0 - E)/A[3]), 2)))
    else:
        res = res + (A[8] * np.exp(-0.5 * np.power(((E_n + 764.0 - E)/A[9]), 2)))
        
    return res

````

````python
## I edited the spreadsheet to trim whitespace and removed uncertainty values
## in the A parameter fields then exported to CSV before reading in.  
data = pd.read_csv("~/Downloads/3He_Response_Parameters.csv", float_precision=None)

num_En = len(data.index) # how many neutron energies we're looking at, the number of rows

E = np.linspace(150.0, 1000.0, num=np.uint32((3000-150) / 5))
E_n = np.zeros(num_En, dtype=np.float64)
A = np.zeros((num_En, 19), dtype=np.float64)

for row in range(0, num_En):
    # first entry in row is the energy in [KeV]
    E_n[row] = data.loc[row][0]
    
    # take slice of everything after and including column "1" (2nd)
    tmp = data.loc[row][1:].to_numpy(dtype=np.float128)
    # insert a leading zero to make subscripts in algo work
    tmp = np.insert(tmp, 0, 0.0)
    # insert a zero column for A10 since it's missing in the data
    tmp = np.insert(tmp, 10, 0.0)
    
    A[row] = tmp

    
resp = np.zeros((num_En, len(E)), dtype=np.float128)
for i in range(0, num_En):
    for j in range(0, len(E)):
        resp[i,j] = response_func(E[j], E_n[i], A[i])
````
