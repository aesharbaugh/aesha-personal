This notebook plots the P-V diagrams for Helium (a noble or "ideal" gas) and Nitrogen (a non-ideal gas) for comparison. Calculating the compressive work to go from 40 atm to 100 atm at a constant 75 degrees Fahrenheit. 

Isothermal data for Helium and Nitrogen aquired from NIST: https://webbook.nist.gov/chemistry/fluid/

```python
#importing relevant libraries

import numpy as np
from numpy import trapz
import matplotlib.pyplot as plt
```

```python
#recording relevant pressure and volume values from NIST website 
#units of volume are mol/l and units of pressure are atm

helium_volume = [0.75272, 0.50572, 0.38222, 0.30810, 0.25868, 0.22338]
helium_pressure = [32.897, 49.346, 65.795, 82.244, 98.692, 115.14]

nitrogen_volume = [0.67055, 0.33597, 0.22596]
nitrogen_pressure = [36.187, 72.374, 108.56]
```

```python
#making P-V diagram using MatPlotLib

plt.plot(helium_volume, helium_pressure, "b", label = "helium")
plt.plot(nitrogen_volume, nitrogen_pressure,"r", label = "nitrogen")

plt.ylim([0, 120])
plt.xlim([0.2, 0.7])
plt.title("P-V Diagram of Helium and Nitrogen")

plt.ylabel('Pressure (atm)')
plt.xlabel('Volume (mol/l)')

plt.legend(loc="upper right")

plt.grid()
plt.show()
```

```python
#calculating compressive work to go from 30 - 100 atm

he_y = np.array(helium_pressure)
ni_y = np.array(nitrogen_pressure)

#estimating change in x to drop from 100 atm to 30 atm

upper = 0.64
lower = 0.25

delta_X = upper-lower

#compute the area under the curves using the composite trapezoidal rule

area = trapz(he_y, dx = delta_X)
print("Compressive Work done on Helium = ", area)

area = trapz(ni_y, dx = delta_X)
print("Compressive Work done on Nitrogen = ", area)
```
